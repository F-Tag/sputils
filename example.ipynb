# Exmaple of ```sputils```

## import dependentencies

````python
from pysptk.util import example_audio_file
from librosa import load
from librosa.display import waveshow, specshow

from IPython.display import Audio, display
````

## load wav file

````python
wave, sr = load(example_audio_file(), sr=16000)

waveshow(wave, sr=sr)
Audio(wave, rate=sr)
````

## remove DC bias and low frequencies

````python
from sputils.preprocess import remove_dc

wave = remove_dc(wave, sr)

waveshow(wave, sr=sr)
Audio(wave, rate=sr)
````

## extract mel-spectrogram

````python
from sputils.feature import melspectrogram

melspec = melspectrogram(wave, sr)
specshow(melspec.T, cmap="jet")
````

## mel-spectrogram to audio

````python
from sputils.feature import mel_to_audio

wave_r = mel_to_audio(melspec, sr)
waveshow(wave_r, sr=sr)
Audio(wave_r, rate=sr)
````
